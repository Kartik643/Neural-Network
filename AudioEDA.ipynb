```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd
%matplotlib inline
```

```python
##reading the sample audio using librosa
audio_file = "/content/85544-3-7-0.wav"
audio, sample_rate = librosa.load(audio_file)
```

```python
print(audio)
```

```python
#plotting the audiondata
plt.figure(figsize=(14, 5))
# librosa.display.waveplot(audio, sr=sample_rate)
plt.plot(audio)
```

```python
# MFCC for feature extraction
# summarises the frequency distribution  across the window size so it is possible to analyze both the frequency and time characteristics of the sound.This audio representation will allow us to identify features for classification
mfcc = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
print(mfcc.shape)
```
